### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
city_data_df.head()
```

```python
ride_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"]).set_index('city')

# Display the data table for preview
pyber_data_df.head(50)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby('type')['ride_id'].count()
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby('type')['driver_count'].sum()
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby('type')['fare'].sum()
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fares / total_rides
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fares / total_drivers
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides" : total_rides,
    "Total Drivers" : total_drivers,
    "Total Fares" : total_fares,
    "Average Fare per Rider" : avg_fare_per_ride,
    "Average Fare per Driver" : avg_fare_per_driver
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_output = pyber_summary_df.copy()
pyber_summary_output['Total Rides'] = pyber_summary_output['Total Rides'].map("{:,}".format)
pyber_summary_output['Total Drivers'] = pyber_summary_output['Total Drivers'].map("{:,}".format)
pyber_summary_output['Total Fares'] = pyber_summary_output['Total Fares'].map("${:,.2f}".format)
pyber_summary_output['Average Fare per Rider'] = pyber_summary_output['Average Fare per Rider'].map("${:,.2f}".format)
pyber_summary_output['Average Fare per Driver'] = pyber_summary_output['Average Fare per Driver'].map("${:,.2f}".format)
pyber_summary_output
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fares_df = pyber_data_df.groupby(by=['type','date'])[['fare']].sum()
total_fares_df
```

```python
total_fares_df_reset = total_fares_df.reset_index()
total_fares_pivot = total_fares_df_reset.pivot(index='date', columns='type', values='fare')
total_fares_pivot.head(50)
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
specific_date_df = total_fares_pivot.loc['2019-01-01':'2019-04-29']
specific_date_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
specific_date_df.index = pd.to_datetime(specific_date_df.index)
specific_date_df.index
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
specific_date_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
date_resampled_df = specific_date_df.resample('W').sum()
date_resampled_df
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

ax = date_resampled_df.plot(
    title="Total Fare by City Type",
    figsize=(20,7), 
)
ax.set_xlabel('')
ax.set_ylabel('Fare ($USD)')
fig = ax.get_figure()
fig.savefig("analysis/PyBer_fare_summary.PNG")
```
