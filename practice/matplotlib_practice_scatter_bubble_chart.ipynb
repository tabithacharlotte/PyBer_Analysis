```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# to create scatter plot in MATLAB, add 'o'
plt.plot(x_axis, y_axis, 'o')
```

```python
#OR You can use scatter function
plt.scatter(y_axis, x_axis, color='red')
plt.xlabel('Fare($)')
plt.title ('PyBer Fare by Month')
plt.ylabel('Date')
plt.xlim(0, 50)
plt.legend(['Chicago'])
plt.gca().invert_yaxis()
```

```python
# bubble chart in MATLAB use 's' or 'size' parameter
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# We can adjust the size by multiplying the data in the y-axis by any value.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# We can refactor the code we used to create a scatter plot above, but instead of using a for loop to create the list y_axis_larger for the size parameter of the marker, we can use Python's list comprehension technique inside the plt.scatter() function. 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# scatterplot Object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# bubble chart object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
# Change the color of the markers to sky blue and make 20 percent tranparent, change the size of the markers to 5 times each data point, edge around circles = black.
ax.scatter(
    y_axis, 
    x_axis, 
    color='skyblue',
    alpha=.8, 
    s =[i * 5 for i in y_axis], 
    edgecolors = 'black',
    linewidth = 2,
)
# Add a title.
ax.set_title('PyBer Fare by Month')
# Invert the y-axis so that January is at the top.
ax.invert_yaxis()
# Switch the axis so that the Fare($) data is on the x-axis.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Add a limit for the x-axis data.
ax.set_xlim(0,50)
# Add a legend for the city of Boston.
ax.legend(['Boston'])
```
