```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# If you're graphing nominal variables (e.g., favorite music groups among teens, number of votes by county or state) where you sort the data from greatest to least or least to greatest, it's best to use a horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# if you want data on opposite axes, switch the arguments in the barh() functin, and then run the cell again
plt.barh(x_axis, y_axis)
# We should invert the y-axis of the previous chart to have "January" at the top and "December" at the bottom because we don't want to show the CEO ride-sharing data with the months in the wrong order!
plt.gca().invert_yaxis()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
#invert
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal plot with ax.plt() to use object-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan')
ax.invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend('Chicago')

##LEGEND TO SAY CHICAGO?
```
